```python
bins = [18, 30, 40, 50, 61]
IkÃ¤ = ['18-29','30-39','40-49','50-']

df['luokkarajat'] = pd.cut(df['Age'], bins, right = False)
df['luokka'] = pd.cut(df['Age'], bins, labels = IkÃ¤, right = False)
pd.crosstab(df['luokkarajat'], 'lkm')
```

```python
df1 = pd.crosstab(df['luokka'], 'lkm', normalize = 'columns')
df1.columns.name = ' '
(df1*100).style.format('{:.1f} %')
```

```python
def get_stats(group):
    return {'min':group.min(), 'max':group.max(),
           'lkm':group.count(), 'keskiarvo':group.mean()}
df2 = df.groupby('luokka')['MonthlyIncome'].apply(get_stats).unstack()
df2 = df2[['lkm', 'min', 'max', 'keskiarvo']]
df2.style.format('{:.1f}')
```
